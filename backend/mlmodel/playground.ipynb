```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data/loan_data.csv")

data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
X = data[["Married", "ApplicantIncome", "Education", "LoanAmount", "Credit_History"]]
y = data[["Loan_Status"]]

X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

y["Loan_Status"] = le.fit_transform(y["Loan_Status"])
X["Married"] = le.fit_transform(X["Married"])
X["Education"] = le.fit_transform(X["Education"])


X
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

X = ss.fit_transform(X)
```

```python
X
```

```python
import joblib

joblib.dump(ss, "Scaler.pkl")
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.metrics import accuracy_score


def model_accuracy(preds):
    return f"Accuracy score: {accuracy_score(y_test, preds)}"
```

```python
from sklearn.model_selection import GridSearchCV, KFold

from xgboost import XGBClassifier

xgb = XGBClassifier()
```

```python
params = {
    "learning_rate": [0.05, 0.20, 0.25],
    "max_depth": [5, 8, 10, 12],
    "min_child_weight": [1, 3, 5, 7],
    "gamma": [0.0, 0.1, 0.2, 0.4],
    "colsample_bytree": [0.3, 0.4, 0.7],
}

gridxgb = GridSearchCV(XGBClassifier(), params, n_jobs=-1, cv=KFold(n_splits=3))

gridxgb.fit(x_train, y_train)
```

```python
preds = gridxgb.predict(x_test)

model_accuracy(preds)
```

```python
gridxgb.best_params_
```

```python
final_model = XGBClassifier(colsample_bytree=0.4, gamma=0.4, learning_rate = 0.05, max_depth=5, min_child_weight=5).fit(x_train, y_train)

preds = final_model.predict(x_test)

model_accuracy(preds)
```

```python
joblib.dump(final_model, "model.pkl")
```
